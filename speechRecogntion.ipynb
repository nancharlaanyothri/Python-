```python
# try the code by keeping your volume up

import pyttsx3
engine =pyttsx3.init()

def speak(text):
    """making the assistent speak"""
    engine.say(text)
    engine.runAndWait()

def main():
    speak("Hello! I am single bot from mallareddy university.")
    speak("You can say hello ,ask my name,or say goodbye.")
    while True:
        command = input("You: ").lower()
        if "hello" in command:
            speak("Hi there! welcome to mallareddy university.")
        elif "what is your name" in command or "whats your name" in command:
            speak("My name is anyothri from mallareddy university.")
        elif "goodbye" in command:
            speak("Goodbye! have a nice day at mallareddy university")
            break
        else:
            speak("I didn't understand that,please try again.")
if __name__ == "__main__":
    main()
```
